# 1. 유아용품 조회 빈도 분석(baby)

## 1) 시간대별 유아용품 조회 빈도 분석

- 30대의 유아용품 조회 빈도가 전체적으로 제일 높다. 그 뒤로는 40대, 20대, 50대, 60대이상 순이다.
- 02-06시, 06-10시 제외하면 조회가 활발하게 이루어진다

```python
time_list = []
time = df_baby['time'].values.tolist()
for i in time:
    if i not in time_list:
        time_list.append(i)
        
age_list = []
age = df_baby['age'].values.tolist()
for j in age:
    if j not in age_list:
        age_list.append(j)        

columns = ['product','time','age','total']
dataset_total=[]

for time in time_list:
    t = df_baby[df_baby['time']==time].copy() # 시간대별로 순서대로 추출
    for age in age_list:
        a = t[t['age']==age].copy()
        sum_list = a['total'].values.tolist()
        sum_list = list(map(int,sum_list))
        sum_total = sum(sum_list)
        dataset_total.append(['유아용품',time,age,sum_total])
        
pd_data = pd.DataFrame(dataset_total,columns=columns)
pd_data = pd_data.sort_values('age')
pd_data = pd_data.reset_index(drop=True)

index = np.array([0,1,2,3,4])

plt.figure(figsize=(14,7))
plt.title('시간대별 유아용품 조회 빈도')

w=0.1
plt.bar(index-(3*w),pd_data[pd_data['time']=='02-06시']['total'],width=w,label='02-06시')
plt.bar(index-(2*w),pd_data[pd_data['time']=='06-10시']['total'],width=w,label='06-10시')
plt.bar(index-(1*w),pd_data[pd_data['time']=='10-14시']['total'],width=w,label='10-14시')
plt.bar(index,pd_data[pd_data['time']=='14-18시']['total'],width=w,label='14-18시')
plt.bar(index+(1*w),pd_data[pd_data['time']=='18-22시']['total'],width=w,label='18-22시')
plt.bar(index+(2*w),pd_data[pd_data['time']=='22-02시']['total'],width=w,label='22-02시')

plt.legend()
plt.xticks(index-(w*0.5),['20대','30대','40대','50대','60대이상'])
plt.grid(axis='y')
plt.savefig('./images/baby_time.png',dpi=200)
plt.show()
```
